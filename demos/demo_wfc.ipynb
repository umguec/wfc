```python
from io import BytesIO
from urllib.request import urlopen

import PIL.Image
import imageio
import numpy as np
from IPython.display import Image

from wfc import WFC
```

```python
input_bitmap = 'https://github.com/mxgmn/WaveFunctionCollapse/raw/master/samples/Village.png'
output_shape = (64, 64)
window_shape = (3, 3)

output_bitmaps = WFC(np.array(PIL.Image.open(BytesIO(urlopen(input_bitmap).read()))),
                     output_shape,
                     window_shape)()

imageio.mimsave('output_bitmaps.gif', output_bitmaps)
Image(open('output_bitmaps.gif', 'rb').read(), width=512, height=512)
```
